```python
import cv2
import numpy as np

def activar_canales(img1, color):
    alto, ancho = img1.shape[:2]
    img1_salida = np.zeros_like(img1)
    
    # Pixel
    for i in range(alto):
        for j in range(ancho):
            pixel = img1[i, j]  
            red = pixel[2]
            green = pixel[1]
            blue = pixel[0]
                
            if color == 1:
                img1_salida[i, j] = [blue, 0, 0]  # Solo azul
            elif color == 2:
                img1_salida[i, j] = [0, green, 0]  # Solo verde
            elif color == 3:
                img1_salida[i, j] = [0, 0, red]  # Solo rojo
            elif color == 10:
                img1_salida[i, j] = [0, green, red]  # Verde y rojo
            elif color == 20:
                img1_salida[i, j] = [blue, green, 0]  # Verde y azul
            elif color == 30:
                img1_salida[i, j] = [blue, 0, red]  # Azul y rojo
    
   
    cv2.imwrite("imagen1_monocromo.jpg", img1_salida)
    return img1_salida

# prueba
img1 = cv2.imread("imagen_problema1.jpg") 
if img1 is None:
    raise FileNotFoundError("No se pudo cargar la imagen. Verifica la ruta.")
    
imagen_procesada = activar_canales(img1, 10)
cv2.imshow("Imagen procesada", imagen_procesada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

¿Que es espacio de color HSV?

Es un modelo que representa los colores de manera mas intuitiva que el RGB, 
basandose en como los humanos perciben los colores.

Los componentes:
-HUE: representa el color puro y  se mide en grados
-Saturacion: Indica la intensidad del color, medido en porcentaje
-Valor: Representa el brillo del color, y es medido en porcentaje
